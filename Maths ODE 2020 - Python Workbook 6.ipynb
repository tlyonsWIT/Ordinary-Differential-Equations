## Step 1 

* In the code cells below we define python functions for the Fourier coefficients $a_{0}$, $a_{n}$ and $b_{n}$.


* The coefficients defined below are those found in __Lecture 11 - Example 4__ i.e.

$$
a_{0}=2\quad a_{n}=0\quad b_{n}=\frac{4}{n\pi}(1-\cos(n\pi)).
$$

```python
a0=2
```

```python
def a(n):
    return 0
```

```python
def b(n):
    return (4/n/pi)*(1-cos(n*pi))
```

## Step 2

* We now define a python function to compute the _partial sum_ of the first $N$ terms of the Fourier series i.e.

$$
f(x,N) = a_{0} +\sum_{n=1}^{N}a_{n}\cos\left(\frac{2\pi n x}{L}\right) + \Sigma_{n=1}^{N}b_{n}\sin\left(\frac{2\pi n x}{L}\right)
$$

```python
def f(x,N,L):
    return a0 + sum(a(n)*cos(2*pi*n*x/L) + b(n)*sin(2*pi*n*x/L) for n in range(1,N+1))
```

## Step 3

* Define an interval for $x$, we will use $0\leq x\leq 3L$ with 500 point, and then plot $f(x,N)$.


* We are going to plot the partial sums of the first 5, 50 and 500 terms.

```python
L=2
X=np.linspace(0,3*L,500)
```

```python
N1=5
N2=50
N3=500
plt.plot(X,f(X,N1,L))
```

```python
plt.plot(X,f(X,N2,L))
```

```python
plt.plot(X,f(X,N3,L))
```
